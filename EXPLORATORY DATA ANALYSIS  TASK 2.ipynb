```python
import pandas as pd

import matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'svg' #Â use to give higher quality plot 
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import seaborn as sns 
```

```python
df = pd.read_csv('data_processed.csv', index_col=0)
```

```python
features = [
    'Channel',
    'Region',
    'Fresh',
    'Milk',
    'Grocery',
    'Frozen',
    'Detergents_Paper',
    'Delicatessen'
]
```

```python
fig, axes = plt.subplots(2, 4, figsize=(7,4))
# because acces is gonna be two dimensional like 2D ARRAY  SO WE DO USE RAVEL : to make multiple PLOT IN SINGLE big FIGURE

for feature, ax in zip(features,axes.ravel()):# ravel : to make plo
    if (feature=='Channel') | (feature=='Region'):
        df[feature].value_counts().plot.bar(ax=ax)
    else:
        ax.hist(df[feature],bins= 100)
    ax.set_title(feature)
plt.tight_layout()  #Â this just helps to keeps things nice and prevents overlapping and unnecessary space at the same time  
    
        
        
        
```

```python
pplot =sns.pairplot(df, vars= features[2:])
pplot.fig.set_size_inches(7,7)
```
